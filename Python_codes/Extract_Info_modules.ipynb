```python
#Testing# importing required modules
import fitz
import csv
from itertools import dropwhile
import numpy as np
import pandas as pd
import PyPDF2
from PyPDF2 import PdfFileMerger, PageRange,PdfWriter, PdfReader

# with open('Modulkatalog_2022_Wintersemester.pdf', 'rb') as file:
#     doc = PyPDF2.PdfF'ileReader(file)


doc = fitz.open('Modulkatalog_2022_Wintersemester.pdf')
text = ""
for page in doc:
    text+=page.get_text()
    
columns_list = ["Modulbezeichnung:","engl. Modulbezeichnung:","ggf. Modulniveau:","Kürzel:","ggf. Untertitel:","ggf. Lehrveranstaltungen:","Studiensemester:","Semesterlage:","Modulverantwortliche(r):","Dozent(in):","Sprache:","Zuordnung zum Curriculum:","Lehrform / SWS:","Arbeitsaufwand:","Kreditpunkte:","Voraussetzungen nach Prüfungsordnung:","Empfohlene Voraussetzungen:","Angestrebte Lernergebnisse:","Inhalt:","Studien-/ Prüfungsleistungen:","Medienformen:","Literatur:"]

text = text.split('\n \n \n \n \nSeite 14   Inhaltsverzeichnis \n     \n \n',1)[1]
```

```python
for pages in range(len(doc)+1, 1, -1):
    text = text.replace("Seite "+ str(pages), " ")
    
text = text.replace("    Inhaltsverzeichnis \n", " ")
text = text.replace("Voraussetzungen nach \nPrüfungsordnung:", "Voraussetzungen nach Prüfungsordnung:")
text_list = list(text.split("\n"))

refine = []
count = 0

for i in range(len(text_list)):
    
    if text_list[i].strip() in columns_list:
        refine.append(text_list[i])
        refine.append("")
    
    else:
        refine[-1] += text_list[i]

    
def group_list(lst):
    data = dict()
    key = 'Modulbezeichnung:'
    for item in lst:
        if item.strip() in columns_list:
            key = item
            if(key not in data.keys()):
                data[key] = list() # saying that the value for data with the specific key is a list
                data[key].append(item)
        if(item != key):
            data[key].append(item)
        
    return list(data.values()) # data.values give as a dict values so cobverting it into list

result = group_list(refine)
df = pd.DataFrame(result)
df = df.T
df = df.rename(columns=df.iloc[0]).loc[1:] #converting first row as column
df
```
